```python
import pandas as pd
import numpy as np
import seaborn as sns
import statistics as sts
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis
from sklearn.preprocessing import StandardScaler
df= pd.read_csv('sales_data_with_discounts.csv')
df.head()
```

```python
mean= df["Volume"].mean()
print("Mean of Volume:", mean)
median= df["Volume"].median()
print("Meadian of Volume:", median)
mode= df["Volume"].mode()
print("Mode of Volume:", mode)
std_dev = sts.stdev(df["Volume"])
print("Standard Deviation of Volume:", std_dev)
```

```python
# Standardization of Numerical Variables
#  	Objective: To scale numerical variables for uniformity, improving the datasets suitability for analytical models.
#  Steps:
#	Explain the concept of standardization (z-score normalization).
#	Standardize the numerical columns using the formula: z=x-mu/sigma	
#	Show before and after comparisons of the data distributions.
```

```python
''' Concept of standardization(z-score normalization)
 ANS: Standardization (also called Z-score normalization) is a data preprocessing technique used to rescale numerical variables.that is 
       Mean=0
       Standard deviation=1
       It helps to make the features comparible and suitable for statistical analysis.We can use StandardScaler to standardize our dataset'''
```

```python
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
print("Numeric columns to standardize:", numeric_cols)
df_scaled=df.copy()
scaler=StandardScaler()
df_scaled[numeric_cols]=scaler.fit_transform(df[numeric_cols])
print("Before Standardization:")
print(df[numeric_cols].describe())
print("\nAfter Standardization:")
print(df_scaled[numeric_cols].describe())
```

```python
'''Conversion of Categorical Data into Dummy Variables
	Objective: To transform categorical variables into a format that can be provided to ML algorithms.
	Steps:
	Discuss the need for converting categorical data into dummy variables (one-hot encoding).
	Apply one-hot encoding to the categorical columns, creating binary (0 or 1) columns for each category.
	Display a portion of the transformed dataset.'''
```

```python
'''Categorical columns need to be converted to numbers because most machine learning models cannot process text.
    Nominal data which has no order we can use one-hot-encoding
    Conversion allows models to interpret categories mathematically and ensure better performance'''
```

```python
categorical_cols = df.select_dtypes(include=['object']).columns
print("Categorical columns:", categorical_cols)
df_encoded = pd.get_dummies(df, columns=categorical_cols, drop_first=True)  # drop_first avoids dummy variable trap
encoded_cat_cols = [col for col in df_encoded.columns if col not in df.select_dtypes(exclude=['object']).columns]
print("Encoded categorical columns:")
print(df_encoded[encoded_cat_cols].head())
```
