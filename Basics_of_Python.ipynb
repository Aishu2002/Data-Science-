```python
# Exercise 1 : Prime Numbers
#Write a Python program that checks whether a given number is prime or not. 
#A prime number is a natural number greater than 1 that has no positive divisors other than 1 and itself.
```

```python
num = float(input("Enter a number : "))
if num<=1:
    print("It is not a Prime number")

if num>1:
    for i in range(2,num):
        if num % i ==0:
            print(num," is not a Prime number") 
            break
    else:
        print(num,"is a Prime number")
```

```python
#Exercise 2: Product of Random Numbers
#Develop a Python program that generates two random numbers and asks the user to enter the product of these numbers. 
#The program should then check if the user's answer is correct and display an appropriate message.
```

```python
#Exercise 3: Squares of Even/Odd Numbers
#Create a Python script that prints the squares of all even or odd numbers within the range of 100 to 200. 
#Choose either even or odd numbers and document your choice in the code.
```

```python
#the squares of all even numbers
for i in range(100,201):
    if i % 2 ==0:
        square = i**2
        print(f"The square of {i} is {square}")

#the squares of all odd numbers
for numbers in range(100,201):
    if numbers % 2 != 0:
        squareODD = numbers**2
        print(f"The square of {numbers} is {squareODD}")
```

```python
#Exercise 4: Word counter
#write a program to count the number of words in a given text.
#example:
#input_text = "This is a sample text. This text will be used to demonstrate the word counter."
```

```python
input_text = "This is a sample text . This text will be used to demonstrate the word counter."

words = input_text.split() #This line breaks the sentence into individual words using spaces as separators.

word_counts = {} #Here, we are creating an empty dictionary.This will be used to store each word and how many times it appears.

# Count each word
for word in words:
    if word in word_counts:
        word_counts[word] += 1
    else:
        word_counts[word] = 1

# Print word counts
for word, count in word_counts.items(): #.items() means "both the key (word) and the value (count) from the dictionary."
    print(f"'{word}': {count}")
```

```python
#Exercise 5: Check for Palindrome
#Write a Python function called is_palindrome that takes a string as input and returns True if the string is a palindrome, 
#and False otherwise. A palindrome is a word, phrase, number, or other sequence of characters that reads the same forward 
#and backward, ignoring spaces, punctuation, and capitalization.
```

```python
s = input("Enter a string")

revstr = (s[::-1])

if s == revstr:
    print("True")
else:
    print("False")
```
