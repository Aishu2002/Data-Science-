```python
'''Assignment Tasks
a. Build 99% Confidence Interval Using Sample Standard Deviation
Assuming the sample is representative of the population, construct a 99% confidence interval for the mean number of characters printed 
before the print-head fails using the sample standard deviation. Explain the steps you take and the rationale behind using the t-distribution 
for this task.'''
```

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Sample data: durability in millions of characters
durability = np.array([1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 
                       1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29])
```

```python

# Step 1: Sample size, mean, and standard deviation
n = len(durability)
mean = np.mean(durability)
std_dev = np.std(durability, ddof=1)  # sample standard deviation

# Step 2: Confidence level and degrees of freedom
confidence = 0.99
df = n - 1
alpha = 1 - confidence

# Step 3: t critical value
t_crit = stats.t.ppf(1 - alpha/2, df)

# Step 4: Standard error
se = std_dev / np.sqrt(n)

# Step 5: Margin of error
margin_error = t_crit * se

# Step 6: Confidence interval
ci_lower = mean - margin_error
ci_upper = mean + margin_error

# Display results
print(f"Sample Mean: {mean:.4f} million characters")
print(f"Sample Standard Deviation: {std_dev:.4f} million characters")
print(f"99% Confidence Interval: ({ci_lower:.4f}, {ci_upper:.4f}) million characters")
```

```python
'''Sample standard deviation (ddof=1) is used because we don’t know the population standard deviation.

t-critical value is obtained from scipy.stats.t.ppf for a two-tailed 99% confidence interval.

Standard error measures how much the sample mean is expected to vary.

Margin of error accounts for variability and confidence level.

Confidence interval gives a range likely containing the true population mean.'''
```

```python
# -------------------------------
# Plotting the mean and confidence interval
# -------------------------------
plt.figure(figsize=(6,4))
plt.errorbar(x=1, y=mean, yerr=margin_error, fmt='o', color='blue', 
             ecolor='red', elinewidth=2, capsize=5, label='99% CI')
plt.xlim(0.5, 1.5)
plt.xticks([])
plt.ylabel("Durability (million characters)")
plt.title("Sample Mean with 99% Confidence Interval")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
'''What the Plot Shows

The blue dot represents the sample mean.

The red error bar represents the 99% confidence interval, extending from ci_lower to ci_upper.

This visual makes it easy to see the range within which the true mean durability is likely to lie with 99% confidence.'''
```

```python
'''b. Build 99% Confidence Interval Using Known Population Standard Deviation
If it were known that the population standard deviation is 0.2 million characters, construct a 99% confidence interval for the mean
number of characters printed before failure.'''
```

```python
# Sample mean and known population standard deviation
mean = 1.2387
sigma = 0.2
n = 15
confidence = 0.99

# Standard error
se = sigma / np.sqrt(n)

# z-critical value for 99% confidence
alpha = 1 - confidence
z_crit = stats.norm.ppf(1 - alpha/2)

# Margin of error
margin_error = z_crit * se

# Confidence interval
ci_lower = mean - margin_error
ci_upper = mean + margin_error

print(f"Sample Mean: {mean:.4f} million characters")
print(f"Population Standard Deviation: {sigma:.4f} million characters")
print(f"99% Confidence Interval (using known sigma): ({ci_lower:.4f}, {ci_upper:.4f}) million characters")
```

```python
'''Uses z-distribution because the population standard deviation (σ) is known.
Two-tailed CI accounts for variability on both sides of the mean.
The confidence interval gives a range that likely contains the true population mean.
This CI is generally narrower than a t-based CI for the same data because knowing σ reduces uncertainty.'''
```
