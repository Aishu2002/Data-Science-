```python
'''Exercise 1: Prime Numbers
Write a Python program that checks whether a given number is prime or not. A prime number is a natural number greater than 1 that has no
positive divisors other than 1 and itself.'''
```

```python
# Function to check if a number is prime
def is_prime(n):
    if n <= 1:
        return False  # Numbers less than or equal to 1 are not prime
    for i in range(2, int(n**0.5) + 1):  # Check divisors up to square root of n
        if n % i == 0:
            return False  # Found a divisor, not prime
    return True  # No divisors found, number is prime

# Input from user
num = int(input("Enter a number to check if it is prime: "))

# Check and display result
if is_prime(num):
    print(f"{num} is a prime number.")
else:
    print(f"{num} is not a prime number.")
```

```python
'''Numbers ≤ 1 are not prime.
To check if a number n is prime, we only need to test divisibility up to the square root of n.
If n has a factor greater than its square root, the corresponding factor must be smaller than the square root.
If no divisors are found, the number is prime.'''
```

```python
'''Exercise 2: Product of Random Numbers
Develop a Python program that generates two random numbers and asks the user to enter the product of these numbers. The program should then
check if the user's answer is correct and display an appropriate message.'''
import random

# Step 1: Generate two random numbers between 1 and 10
num1 = random.randint(1, 10)
num2 = random.randint(1, 10)

# Step 2: Ask the user for the product
user_input = int(input(f"What is the product of {num1} and {num2}? "))

# Step 3: Check the answer
correct_product = num1 * num2

if user_input == correct_product:
    print("Correct! Well done.")
else:
    print(f"Incorrect. The correct answer is {correct_product}.")

```

```python
'''Exercise 3: Squares of Even/Odd Numbers
Create a Python script that prints the squares of all even or odd numbers within the range of 100 to 200. Choose either even or odd numbers
and document your choice in the code.'''

# Exercise 3: Squares of Even Numbers from 100 to 200
# We choose even numbers in the range 100 to 200
print("Squares of even numbers from 100 to 200:")

for num in range(100, 201):  # 201 because the upper bound is exclusive
    if num % 2 == 0:  # Check if the number is even
        square = num ** 2
        print(f"{num}^2 = {square}")
```

```python
#Exercise 4: Word counter
#write a program to count the number of words in a given text.

# Input text
input_text = "This is a sample text. This text will be used to demonstrate the word counter."

# Convert the text to lowercase (optional, remove if case-sensitive count is desired)
# input_text = input_text.lower()

# Split the text into words
words = input_text.split()

# Create a dictionary to store word counts
word_count = {}

# Count the occurrences of each word
for word in words:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1

# Print the word counts
for word, count in word_count.items():
    print(f"'{word}': {count}")
```

```python
'''Exercise 5: Check for Palindrome
Write a Python function called is_palindrome that takes a string as input and returns True if the string is a palindrome, 
and False otherwise. A palindrome is a word, phrase, number, or other sequence of characters that reads the same forward and backward, 
ignoring spaces, punctuation, and capitalization.'''

import string

# Function to check if a string is a palindrome
def is_palindrome(s):
    # Convert to lowercase
    s = s.lower()
    
    # Remove spaces and punctuation
    s = ''.join(char for char in s if char.isalnum())
    
    # Check if the string reads the same forwards and backwards
    return s == s[::-1]

# Example usage
input_text = "racecar"
print(is_palindrome(input_text))  # Output: True

# Another example with spaces and punctuation
input_text2 = "A man, a plan, a canal, Panama!"
print(is_palindrome(input_text2))  # Output: True
```

```python
'''s.lower() → Converts all letters to lowercase to ignore capitalization.
char.isalnum() → Keeps only letters and numbers, removing spaces and punctuation.
s[::-1] → Reverses the string.
The function returns True if the cleaned string equals its reverse, otherwise False.'''
```
